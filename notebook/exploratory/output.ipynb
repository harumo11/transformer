```python
import pickle as pkl
import os
import torch
```

```python
with open('../../data/external/output.pkl', 'rb') as f:
    output_data = pkl.load(f)
```

```python
output_data.shape # 11010 is the kind of the english words
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot output_data with bar chart
plt.figure(figsize=(200, 6))
out = output_data.to('cpu').detach().numpy().flatten()
plt.bar(range(len(out)), out)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Bar Chart of Output Data')
plt.savefig('output_data_bar_chart.png')
plt.show()
```
