```python
import torch
from torch.utils.data import Dataset
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
```

```python
data_dir_path = '../../data/raw/fashionmnist/'
training_data = datasets.FashionMNIST(
    root=data_dir_path,
    train=True, # training set or test set
    download=True, # download if not available
    transform=ToTensor() # specify the feature and label ransformation
)
```

```python
test_data = datasets.FashionMNIST(
    root=data_dir_path,
    train=False, # training set or test set
    download=True, # download if not available
    transform=ToTensor() # specify the feature and label ransformation
)
```

# iterating and visualizaing the dataset

```python
training_data.class_to_idx
```

```python
fig = plt.figure(figsize=(10, 10))
cols, rows = 3, 3
for i in range(1, cols * rows + 1):
    sample_idx = torch.randint(0, len(training_data), size=(1,)).item()
    img, label = training_data[sample_idx]
    fig.add_subplot(rows, cols, i)
    plt.title(training_data.classes[label])
    plt.axis('off')
    plt.imshow(img.squeeze(), cmap='gray')
plt.show()
```
