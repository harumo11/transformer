```python
import spacy
import sys
import os
```

```python
# load dataset 
script_abs_path = '/home/sasatake/repos/python/transformer/script'
sys.path.append(script_abs_path)

import load_dataset
multi30 = load_dataset.Multi30k()
```

```python
# the trial for tokenize of German
nlp_de = spacy.load('de_core_news_sm')
for text in multi30.datasets['texts_src_de_train'][:5]:
    print('-----------------------------------')
    print(text)
    doc = nlp_de(text)

    print('token text \t token pos \t token dep')
    for token in doc:
        print(f'{token.text} \t\t {token.pos_} \t\t {token.dep_}')
```
